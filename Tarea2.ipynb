## 1. Introducci√≥n

Se escogi√≥ el problema de clasificaci√≥n de hongos, "Mushroom Classification", el cual consiste en clasificar de manera binaria si un hongo es comestible o venenoso. Esta etiqueta se encuentra en la columna `class`.

Se cuenta con un dataset de 8.124 observaciones distintas de especies de hongos, identificando 22 cualidades principales, describiendo entre otras cosas el tama√±o o color del p√≠leo, del tallo, etc.

Para conseguir clasificar adecuadamente se implementaron tres modelos matem√°ticos; un modelo Bayesiano, Random Forest y regresi√≥n log√≠stica. Se seleccionaron pertinentemente las caracter√≠sticas principales para iterar y conseguir un mejor resultado. La evaluaci√≥n se realiz√≥ en base a la m√©trica F1 (macro), ya que permite medir ponderadamente el desempe√±o de los verdaderos positivos (precision) junto con la clasificaci√≥n de verdaderos negativos (recall), lo cual es sumamente importante ya que estamos prediciendo algo que podr√≠a ser peligroso en caso de cometer un error, especialmente en la clase positiva de venenoso.

Nuestro modelo cumpli√≥ a la perfecci√≥n las expectativas, y de hecho las rebas√≥, ya que logr√≥ un excelente desempe√±o clasificando hongos comestibles de los venenosos.

A continuaci√≥n se presenta una descripci√≥n de cada una de las variables.

- `cap-shape`: se refiere a la forma de tapa que tiene el hongo, que puede ser campana, c√≥nico, convexo, plano, nudoso o hundido.

- `cap-surface`: se refiere a la superficie de la tapa que tiene el hongo, que puede ser fibroso, surcos, escamoso o liso.

- `cap-color`: se refiere al color de la tapa que tiene el hongo, que puede ser caf√©, ante, canela, gris, verde, rosado, morado, rojo, blanco o amarillo.

- `bruises`: se refiere a si el hongo tiene o no magulladuras, esto se representa de manera binaria con t (true) y f (false).

- `odor`: se refiere al olor que tiene el hongo en diferentes categor√≠as, como almendra, an√≠s, creosota, pescado, mal olor, rancio, ninguno, acre o picante.

- `gill-attachment`: se refiere al apego branquial que tiene el hongo en diferentes categor√≠as, como adjunto, descendente, libre o mellado.

- `gill-spacing`: se refiere al espaciamiento branquial que tiene el hongo en diferentes categor√≠as, como cerrado, atestado o distante.

- `gill-size`: se refiere al tama√±o de las branquias que tiene el hongo en diferentes categor√≠as, como amplio o estrecho.

- `gill-color`: se refiere al color de las branquias que tiene el hongo, como negro, caf√©, ante, chocolate, gris, verde, narango, rosado, morado, rojo, blanco o amarillo.

- `stalk-shape`: se refiere a la forma del tallo que tiene el hongo, como agrandado o estrecho. 

- `stalk-root`: se refiere a la ra√≠z del tallo que tiene el hongo, como bulbosa, club, copa, igual, rizomorfos, enraizado o ninguno.

- `stalk-surface-above-ring`: se refiere a la superficie del tallo sobre el anillo del hongo, como fibroso, escamosa, sedosa o lisa.

- `stalk-surface-below-ring`: se refiere a la superficie del tallo bajo el anillo del hongo, como fibroso, escamosa, sedosa o lisa.

- `stalk-color-above-ring`: se refiere al color del tallo sobre el anillo del hongo, como caf√©, ante, canela, gris, verde, rosado, morado, rojo, blanco o amarillo.

- `stalk-color-below-ring`: se refiere al color del tallo bajo el anillo del hongo, como caf√©, ante, canela, gris, verde, rosado, morado, rojo, blanco o amarillo.

- `veil-type`: se refiere al tipo de velo que tiene el hongo, que puede ser parcial o universal.

- `veil-color`: se refiere al color del velo que tiene el hongo, que puede ser marr√≥n, naranjo, blanco o amarillo.

- `ring-number`: se refiere al n√∫mero de timbre que tiene el hongo, que puede ser ninguno, uno o dos, donde a priori no existe relaci√≥n ordinal y por eso se escriben como palabras.

- `ring-type`: se refiere al tipo de timbre que tiene el hongo, que puede ser telara√±a, evanescente, abocinado, grande, ninguna, colgante, revestimiento o zona.

- `spore-print-color`: se refiere al color de impresi√≥n de esporas, que puede ser negro, marr√≥n, ante, chocolate, verde, naranja, morado, blanco o amarillo.

- `population`: se refiere a la poblaci√≥n en la que vive el hongo, que puede ser abundante, agrupada, numerosa, dispersa, varias o solitaria.

- `habitat`: se refiere al habitat en el que vive el hongo, que pueden ser pastos, hojas, prados, caminos, urbano, desechos o bosques.

## 3. Preparaci√≥n de Datos


En primer lugar se convierte la etiqueta `class` a una binaria, en vez de una categ√≥rica con strings

```python
df['class'] = df['class'].replace({"p" : 1, "e" : 0}) 
X = df.iloc[:, 1:]
y = df.iloc[:, 0]
```

En adelante, `class` vale 1 si el hongo es venenoso, y 0 si no.

Se crean conjuntos de entrenamiento y testeo, con una raz√≥n de $30\%$ para 
los datos de testeo, y el resto del dataset dedicado para el entrenamiento.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
print("La estructura del conjunto de entrenamiento:",X_train.shape)
print("La estructura del conjunto de testeo:",X_test.shape)
```

A continuaci√≥n creamos un ColumnTransformer donde se tiene un OneHotEncoder() para todas las columnas explicativas. Es importante destacar que no existe un orden espec√≠fico en las categor√≠as.


```python
preprocessing_transformer = ColumnTransformer(
    transformers=[
        ('OneHotEncoder', OneHotEncoder(), ['cap-shape', 'cap-surface', 'cap-color', 'bruises', 'odor',
       'gill-attachment', 'gill-spacing', 'gill-size', 'gill-color',
       'stalk-shape', 'stalk-root', 'stalk-surface-above-ring',
       'stalk-surface-below-ring', 'stalk-color-above-ring',
       'stalk-color-below-ring', 'veil-type', 'veil-color', 'ring-number',
       'ring-type', 'spore-print-color', 'population', 'habitat'])], sparse_threshold = 0)
```

---

## 4. Modelo inicial

Adjuntamos las transformaciones anteriores en un solo `Pipeline` y agregamos al final un clasificador sencillo `Naive Bayes`. 

```python
pipeline_inicial = Pipeline([('Preprocessing', preprocessing_transformer), ('NaiveBayes', GaussianNB())])
```

```python
# Se procede a entrenar el pipeline con los datos de entrenamiento y las clases.
pipeline_inicial.fit(X_train, y_train)
```

```python
# Se procede a predecir sobre la data de testeo con el pipeline previamente entrenado.
y_pred = pipeline_inicial.predict(X_test)
```

A continuaci√≥n, se muestran los resultados obtenidos de la predicci√≥n sobre nuestra data de testeo:

```python
print(classification_report(y_test, y_pred))
```

## 5. Optimizaci√≥n del modelo

Ahora, creamos una versi√≥n optimizada del modelo anterior, donde crearemos un pipeline que realice la transformaci√≥n, seleccione caracter√≠sticas y clasifique.

```python
pipeline_optimizado = Pipeline([('Preprocessing', preprocessing_transformer),
                                ('SelectFromModel', SelectFromModel(LogisticRegression())),
                                ('RandomForest', RandomForestClassifier())])
```

```python
# Se procede a entrenar el pipeline con los datos de entrenamiento y las clases.
pipeline_optimizado.fit(X_train, y_train)
```

```python
# Se procede a predecir sobre la data de testeo con el pipeline previamente entrenado.
y_pred2 = pipeline_optimizado.predict(X_test)
```

A continuaci√≥n, se muestran los resultados obtenidos de la predicci√≥n sobre nuestra data de testeo:

```python
print(classification_report(y_test, y_pred2))
```

A continuaci√≥n, se muestran las caracter√≠sticas que selecciona "SelectFromModel" que son las m√°s importantes para la predicci√≥n:

```python
support = pipeline_optimizado.named_steps['SelectFromModel'].get_support()
feature_names = np.array(pd.get_dummies(X_train).columns) 
feature_names[support]
```

Dado que se encontraron resultados muy buenos en el anterior pipeline, encontramos que no era necesario seguir complejizando el modelo, por esto creamos otro pipeline con un clasificador diferente para observar si estos resultados se repet√≠an.

```python
pipeline_optimizado2 = Pipeline([('Preprocessing', preprocessing_transformer),
                                ('SelectFromModel', SelectFromModel(LogisticRegression())),
                                ('LogisticRegression', LogisticRegression())])
```

```python
# Se procede a entrenar el pipeline con los datos de entrenamiento y las clases.
pipeline_optimizado2.fit(X_train, y_train)
```

```python
# Se procede a predecir sobre la data de testeo con el pipeline previamente entrenado.
y_pred3 = pipeline_optimizado2.predict(X_test)
```

A continuaci√≥n, se muestran los resultados obtenidos de la predicci√≥n sobre nuestra data de testeo:

```python
print(classification_report(y_test, y_pred3))
```
