````python
# Creamos el dataframe de temperaturas donde haremos los cambios necesarios para obtener lo que se pide
temperaturas = temp_chile2.drop(['codigo_estacion', 'altura', 'latitud', 'longitud','institucion', 'fuente', 'codigo_cuenca', 'nombre_cuenca', 
                                   'codigo_sub_cuenca', 'nombre_sub_cuenca', 'inicio_observaciones', 'fin_observaciones', 'cantidad_observaciones', 
                                   'inicio_automatica'], axis=1)   # eliminamos las columnas que no necesitaremos y dejamos la estaci√≥n + fechas
temperaturas.index = temperaturas['nombre'] # Asignamos por comodidad el nombre de la estaci√≥n al √≠ndice
temperaturas = temperaturas.drop(['nombre'], axis=1) # Eliminamos la columna nombre
temperaturas.head(3)
````

````python
# Utilizamos melt para pivotear el set de datos
temperaturas = temperaturas.melt(ignore_index = False).reset_index() # ignore_index = False requiere actualizar pd
temperaturas.columns = ['station', 'Dates', 'Temp']
temperaturas.head(3)
````

````python
# Ya que utilizaremos la temperatura promedio por mes, para evitar sesgo vamos a ignorar los d√≠as con temperaturas que no tienen l√≥gica.
temperaturas = temperaturas[ temperaturas['Temp'] != '-9999'] # Primera forma en que est√°n escritas las temperaturas erroneas o c√≥digo para mediciones err√≥neas '-9999'
temperaturas = temperaturas[ temperaturas['Temp'] != -9999].reset_index(drop=True) # Segunda forma en que est√°n escritas, -9999
temperaturas['Year'] = pd.DatetimeIndex(temperaturas['Dates']).year # Creamos la columna con el a√±o al que corresponde la medici√≥n.
temperaturas['Month'] = pd.DatetimeIndex(temperaturas['Dates']).month # Creamos la columna con el mes al que corresponde la medici√≥n.
temperaturas = temperaturas.drop(['Dates'], axis=1) # Eliminamos la columna fecha que ya no vamos a necesitar, pero m√°s adelante crearemos la solicitada.
temperaturas['Temp'] = pd.to_numeric(temperaturas['Temp']) # La serie era de tipo objeto por lo que es necesario actualizar  a num√©rica.
temperaturas_final = temperaturas.groupby(['station','Year', 'Month']).mean().reset_index() # Se realiza un groupby para obtener la temperatura promedio por las variables dadas
temperaturas_mean = temperaturas.groupby(['Year', 'Month']).mean().reset_index()
````

````python
temperaturas_final = temperaturas_final.drop(['Temp'], axis=1)
temperaturas = pd.merge(left=temperaturas_final, right = temperaturas_mean, on = ['Year', 'Month'], how = "inner")
````

````python
#########################
info_station = temp_chile2.loc[:,['nombre', 'altura', 'latitud', 'longitud']] 
info_station.columns = ['station', 'altura', 'latitud', 'longitud']
info_station.head(3)
````

````python
#########################
temp_chile_ord = pd.merge(left=temperaturas, right=info_station, on = ['station'], how="inner") # Unimos las temperaturas con los datos de la estaci√≥n.
#########################
temp_chile_ord['Dates'] = 0
for index in range(len(temp_chile_ord)): # Creamos un loop para poder agregar la nueva columna Dates en base al a√±o y mes.
  una_fecha = str(temp_chile_ord['Year'][index]) + '/' + str(temp_chile_ord['Month'][index]) + '/01 00:00:00'
  temp_chile_ord['Dates'][index] = datetime.datetime.strptime(una_fecha, '%Y/%m/%d %H:%M:%S')
#########################
cols = ['Year', 'Month', 'Dates', 'altura', 'latitud', 'longitud', 'station', 'Temp'] # Orden en el que queremos el dataframe
temp_chile_ord = temp_chile_ord[cols]
temp_chile_ord=temp_chile_ord.sort_values(by=['Dates']) # Se ordena el dataframe en funci√≥n de las fechas
temp_chile_ord=temp_chile_ord.reset_index(drop=True)
months=temp_chile_ord.copy()
temp_chile_ord['Month'] = temp_chile_ord['Month'].apply(lambda x: calendar.month_name[x]) # Cambiamos el mes de manera numerica al nombre correspondiente.
````
